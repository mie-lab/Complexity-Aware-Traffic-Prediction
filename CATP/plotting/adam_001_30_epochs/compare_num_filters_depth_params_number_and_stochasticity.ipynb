```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

# Define a colormap
cmap = plt.get_cmap('viridis')

for p_h in range(1, 10):
    data_all = pd.read_csv("record_distribution_ph"+str(p_h)+".csv", header=None)
    filtered_PM_rows = data_all.iloc[15, :]  # Assuming you want the 16th row
    assert (filtered_PM_rows[:].T.iloc[0] == "MP")
    data = (np.array(filtered_PM_rows.T.iloc[1:]).flatten())
    data = data.tolist()
    print(np.mean(data))

    random.shuffle(data)
    x = []
    y = []
    std_devs = []  # List to hold the standard deviations
    for N in range(10, 1000, 10):
        x.append(N)
        sample_data = data[:N]
        sample_mean = np.mean(sample_data)
        y.append(sample_mean)
        std_devs.append(np.std(sample_data))  # Compute standard deviation for the sample

    # Normalize p_h value to the range of the colormap
    normalized_ph = (p_h - 1) / (9 - 1)
    color = cmap(normalized_ph)

    plt.plot(x, y, label=r'$p_h:' + str(p_h) + '$', color=color)
    plt.fill_between(x, np.subtract(y, std_devs), np.add(y, std_devs), color=color, alpha=0.2)

plt.legend(ncol=3, fontsize=11)
plt.xlabel("Sample size (n) used to compute MC(f) at epoch 15")
plt.ylabel(r"Computed IC= $\sum_{k=1}^{n} {CRIT (x_k)}$")
plt.title(r"MC values for $Task(i_o=4, p_h=*, s=55, city=London)$")
plt.savefig("MC_sensitivity_MonteCarlo.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data Preparation
previous_data_dict = {
    "DEP": [1, 1, 1, 2, 2, 2, 4, 4, 4],
    "FIL": [16, 64, 128, 16, 64, 128, 16, 64, 128],
    "Mean": [57.85, 58.54, 150.25, 34.24, 127.93, 223.59, 146.59, 340.81, 538.14]
}

new_data_dict = {
    "DEP": [1, 2, 4, 1, 2, 4, 1, 2, 4],
    "FIL": [32, 32, 32, 64, 64, 64, 128, 128, 128],
    "Params": [39073, 112993, 260833, 151873, 447169, 1037761, 598657, 1779073, 4139905]
}

df = pd.merge(pd.DataFrame(previous_data_dict), pd.DataFrame(new_data_dict), on=["DEP", "FIL"], how='outer').fillna(0)

# Normalize the Mean and Params values for circle size
df['Mean_norm'] = df['Mean'] / df['Mean'].max() * 500
df['Params_norm'] = df['Params'] / df['Params'].max() * 500

# Plotting
fig, ax = plt.subplots()

for i in range(len(df)):
    # Circle for Mean
    ax.scatter(df['DEP'][i] - 0.1, df['FIL'][i], s=df['Mean_norm'][i], color='green', label='Mean' if i == 0 else "")
    # Circle for Params
    ax.scatter(df['DEP'][i] + 0.1, df['FIL'][i], s=df['Params_norm'][i], color='blue', label='Params' if i == 0 else "")

# Labels and title
ax.set_xlabel('Depth (DEP)')
ax.set_ylabel('# Filters (FIL)')
ax.set_title('Mean and Params values by DEP and FIL')
ax.grid(True)

# Legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2], loc='upper left')  # Only show one legend entry per label

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data Preparation
previous_data_dict = {
    "DEP": [1, 1, 1, 2, 2, 2, 4, 4, 4],
    "FIL": [16, 64, 128, 16, 64, 128, 16, 64, 128],
    "Mean": [71.36, 60.40, 100.67, 33.68, 118.87, 241.92, 151.42, 355.84, 312.60]
}

new_data_dict = {
    "DEP": [1, 2, 4, 1, 2, 4, 1, 2, 4],
    "FIL": [32, 32, 32, 64, 64, 64, 128, 128, 128],
    "Params": [39073, 112993, 260833, 151873, 447169, 1037761, 598657, 1779073, 4139905]
}

df = pd.merge(pd.DataFrame(previous_data_dict), pd.DataFrame(new_data_dict), on=["DEP", "FIL"], how='outer').fillna(0)

# Sort DataFrame by Params
df = df.sort_values(by='Params')

# Create DEP-FIL combinations for x-axis
df['DEP-FIL'] = df.apply(lambda row: f'{row.DEP}-{row.FIL}', axis=1)

# Plotting
fig, ax = plt.subplots()

# Bar widths
width = 0.35

# Bar positions
x = range(len(df))

# Bar plot for #Params
ax.bar([i - width/2 for i in x], df['Params']/10000, width, label=r'$\frac{1}{10000}$ #Params')

# Bar plot for Mean
ax.bar([i + width/2 for i in x], df['Mean'], width, label='Mean MC for \n 3 epochs with lowest Val MSE')

# Labels and title
ax.set_xlabel('DEP-FIL Combination')
ax.set_ylabel('Values')
ax.set_title('Comparison of #Params and Mean by DEP-FIL Combination')
ax.set_xticks(x)

def format_label(row, full_format_indices):
    if row.name in full_format_indices:
        return f'(DEP={row.DEP}, FIL={row.FIL})'
    return f'({row.DEP}, {row.FIL})'

# Specify indices where full format should be used
full_format_indices = [0, len(df) - 1]  # First and last index for example

# Apply the function to create the labels
df['DEP-FIL'] = df.apply(lambda row: format_label(row, full_format_indices), axis=1)

# ...

ax.set_xticklabels(df['DEP-FIL'], rotation=45, ha="right")  # Adjusted rotation and alignment for better readability
ax.legend()

plt.tight_layout()
plt.savefig("num_filters.png", dpi=300)
plt.show()
```
