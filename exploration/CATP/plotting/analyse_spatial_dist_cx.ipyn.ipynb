```python
import numpy as np
import os
import sys
import glob

sys.path.append('../')

import config
import matplotlib.pyplot as plt

fileprefix_list = ["london-4-1-25-", "london-4-1-55-" , "london-4-1-105-", "madrid-4-1-25-", "madrid-4-1-55-", 
                "melbourne-4-1-25-", "melbourne-4-1-55-"]

for prefix in fileprefix_list:
    filenames = glob.glob(os.path.join(os.getcwd(), 
                                       "Spatial_results_multiple_runs", prefix, "_PM_spatial_complexity*.npy"))
    a = 0
    for npy in filenames:
        a = a + np.load(npy)
    max_ = np.max(a)
    plt.hist(a.flatten())
    plt.show()
    plt.imshow(a/max_)
    
    plt.title(prefix)
    plt.colorbar()
    plt.show()
```
