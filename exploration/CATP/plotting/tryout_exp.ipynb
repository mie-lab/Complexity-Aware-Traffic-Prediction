```python
from smartprint import smartprint as sprint

import numpy as np
import matplotlib.pyplot as plt

a = np.random.normal(100, 20, (1000,1))

plt.hist(a, 20)
plt.show()
a = a.flatten().tolist()

sum_exp = 0
exp_sum = 0
for i in range(100):
    x_i = a[i]
    y_i = a[i+1]
    
    sum_exp += np.exp((y_i - x_i)/y_i)
    exp_sum += (y_i - x_i)/y_i

    # sprint (sum_exp)
    # sprint (np.exp(exp_sum))
    
    plt.scatter(i, (sum_exp/(i+1)), color="red")
    plt.scatter(i, np.exp(exp_sum)/(i+1), color="blue")
    
plt.ylabel("Sum exp; red, exp_sum; blue")
plt.xlabel("N")
# plt.yscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
a = 1.0  # Amplitude
b = 1.0  # Frequency

# Define the range of h
h_range = np.linspace(0, 2 * np.pi / np.abs(b), 100)

for t in [1,2,4,100,200]:
    diff = []
    for h in h_range:
        # Calculate the absolute difference for each value of h
        diff.append( np.abs(a * abs(np.sin(b*0)) - a * abs(np.sin(b * (0+h )))) )

    # Plot the variation
    plt.plot(h_range, diff, label="t " + str(t))
    
plt.xlabel('h')
plt.ylabel('|X(t) - X(t+h)|')
plt.title('Variation of |X(t) - X(t+h)| with respect to h')
plt.grid(True)
plt.xlim()
plt.legend()
plt.show()
```
