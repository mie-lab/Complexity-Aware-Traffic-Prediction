$$e^(\frac{\sum{y_i} - \sum{x_i}}{\sum{y_i}})$$
$$\sum{e^(\frac{{y_i} - {x_i}}{{y_i}})}$$

```python
from smartprint import smartprint as sprint

import numpy as np
import matplotlib.pyplot as plt

a = np.random.normal(100, 20, (1000,1))

plt.hist(a, 20)
plt.show()
a = a.flatten().tolist()

sum_exp = 0
exp_sum = 0
for i in range(100):
    x_i = a[i]
    y_i = a[i+1]
    
    sum_exp += np.exp((y_i - x_i)/y_i)
    exp_sum += (y_i - x_i)/y_i

    # sprint (sum_exp)
    # sprint (np.exp(exp_sum))
    
    plt.scatter(i, (sum_exp/(i+1)), color="red")
    plt.scatter(i, np.exp(exp_sum)/(i+1), color="blue")
    
plt.ylabel("Sum exp; red, exp_sum; blue")
plt.xlabel("N")
# plt.yscale('log')
plt.show()
```
