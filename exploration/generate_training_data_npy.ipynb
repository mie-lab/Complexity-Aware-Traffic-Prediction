```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:





# In[1]:


import os
import sys
sys.path.append("../")

! pip install haversine
# ! pip install tensorflow 
# ! pip install smartprint -
! pip install tqdm 
# ! pip install matplotlib
! mkdir data_samples

# In[3]:


import datetime

import matplotlib.pyplot as plt
import numpy as np
import pandas
from IPython.core.display import HTML
from IPython.display import display
from matplotlib.patches import Rectangle
import t4c22
from t4c22.t4c22_config import load_basedir


# In[4]:


get_ipython().run_line_magic('load_ext', 'autoreload')
get_ipython().run_line_magic('load_ext', 'time')
get_ipython().run_line_magic('autoreload', '2')
get_ipython().run_line_magic('autosave', '60')
get_ipython().run_line_magic('matplotlib', 'inline')

display(HTML("<style>.container { width:80% !important; }</style>"))


# In[5]:


# load BASEDIR from file, change to your data root
BASEDIR = load_basedir(fn="t4c22_config.json", pkg=t4c22)


# In[6]:


city = "london"


# In[7]:


BASEDIR


# ## Nodes with counters and corresponding daily assigned counts

# In[8]:


nodes = pandas.read_parquet(BASEDIR / "road_graph" / city / "road_graph_nodes.parquet")
counters = nodes[nodes["counter_info"] != ""]
counters


# In[9]:


daily_counts = pandas.read_parquet(BASEDIR / "loop_counter" / city / "counters_daily_by_node.parquet")
daily_counts = daily_counts.reset_index()
daily_counts['counter_info'] = daily_counts['counter_info'].apply(lambda x: x[0])


nodes['counter_info'] = nodes['counter_info'].astype(str)
daily_counts['counter_info'] = daily_counts['counter_info'].astype(str)


# In[ ]:





# In[ ]:





# In[ ]:





# In[10]:


nodes.dtypes


# In[11]:


# daily_counts2['counter_info'] = daily_counts2['counter_info'].astype(str)
# nodes['counter_info'] = nodes['counter_info'].astype(str)


# In[12]:


import pandas as pd
df = pd.merge(daily_counts, counters, on=["node_id", "counter_info"])
df.shape


# In[13]:


df[df["day"] == "2020-01-01"].head()


# In[ ]:





# In[ ]:





# In[17]:


import os
import sys


sys.path.insert(0, os.path.abspath("../"))  # noqa:E402

from haversine import haversine, Unit
import numpy as np
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas
from IPython.core.display import HTML
from IPython.display import display
from matplotlib.patches import Rectangle
import t4c22
from t4c22.t4c22_config import load_basedir

BASEDIR = load_basedir(fn="t4c22_config.json", pkg=t4c22)

city = "london"

nodes = pandas.read_parquet(BASEDIR / "road_graph" / city / "road_graph_nodes.parquet")
counters = nodes[nodes["counter_info"] != ""]

daily_counts = pandas.read_parquet(BASEDIR / "loop_counter" / city / "counters_daily_by_node.parquet")
daily_counts = daily_counts.reset_index()
daily_counts['counter_info'] = daily_counts['counter_info'].apply(lambda x: x[0])

nodes['counter_info'] = nodes['counter_info'].astype(str)
daily_counts['counter_info'] = daily_counts['counter_info'].astype(str)


df = pd.merge(daily_counts, counters, on=["node_id", "counter_info"])




def get_NS_EW_distance(df):
    # compute the distance between the east-most and west-most points
    east_most = df['x'].max()
    west_most = df['x'].min()
    EW = haversine((east_most, 0), (west_most, 0), unit=Unit.KILOMETERS)

    # compute the distance between the north-most and south-most points
    north_most = df['y'].max()
    south_most = df['y'].min()
    NS = haversine((0, north_most), (0, south_most), unit=Unit.KILOMETERS)
    
from tqdm import tqdm
def convert_df_to_nxnx96(df, n, date_):
#     print (get_NS_EW_distance(df))
    df = pd.DataFrame(df[df["day"] == date_])
    if df.shape[0] < 10: 
        return
        
    x = []
    y = []
    val = []
    for i in (range(df.shape[0])):  # , desc="creating_df"):
#         if i > 10000:
#             continue
        x.append(df.iloc[i].x)
        y.append(df.iloc[i].y)        
        val.append(df.iloc[i].volume)

    data = {'x': x, 'y': y, 'volume': val}

    df = pd.DataFrame(data)
    
    df = df.fillna(0)
    
    get_NS_EW_distance(df)

    # compute the x and y bin edges
    x_bin_edges = np.linspace(df['x'].min(), df['x'].max(), n )
    y_bin_edges = np.linspace(df['y'].min(), df['y'].max(), n )

    # compute the x and y bin indices for each point
    x_bins = np.digitize(df['x'], x_bin_edges) - 1
    y_bins = np.digitize(df['y'], y_bin_edges) - 1

    # create an empty matrix M
    M = np.zeros((n, n, 96))

    # populate the matrix M with the values from the dataframe
    for i in (range(df.shape[0])): # , desc="Creating Matrix"):
        M[x_bins[i], y_bins[i], :] = df['volume'][i]

        
    M = np.nan_to_num(M, 0)
    np.save("data_samples/" + date_ + "-"+str(n) + "x" + str(n) + "x96.npy", M)
#     print(M.shape)

    return M


# n = 10
# M = convert_df_to_nxnx96(df, n, date)


# plt.imshow(M[:, :, 8])
# plt.show()




import datetime
from smartprint import smartprint as sprint 

end_date = datetime.date(2020, 1, 31)
delta = datetime.timedelta(days=1)

dates = []
for n in tqdm([1, 8, 16, 32, 64, 128, 256]):
#     n = 32 

    start_date = datetime.date(2019, 7, 1)
    for i in tqdm(range(220), desc="days processed: "):
#         sprint (start_date, end_date)
        if start_date <= end_date:
            dates.append(start_date.strftime('%Y-%m-%d'))
            start_date += delta
#             print (dates[-1])

            M = convert_df_to_nxnx96(df, n, dates[-1])        
        else:
            break

    print(dates)

    mask = np.random.rand(M.shape[0], M.shape[1]) * 0
    mask[np.sum(M, axis=2)>0]  = 1 
    np.save("data_samples/M.npy", M)

    plt.imshow(mask)


# In[ ]:




```
